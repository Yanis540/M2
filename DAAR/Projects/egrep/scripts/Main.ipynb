```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns 
import pandas as pd 
```

# Accuracy 

```python
#Cr√©ation du DataFrame
df = pd.read_csv('../results/accuracy_and_time.csv')

# Matrices de confusion pour KMP et DFA
confusion_kmp = confusion_matrix(df['found_kmp'], df['appears_expected'])
confusion_dfa = confusion_matrix(df['found_dfa'], df['appears_expected'])
```

```python
# Fonction pour afficher les matrices de confusion
def plot_confusion_matrix(conf_matrix, method_name):
    plt.figure(figsize=(6, 4))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=['False', 'True'], yticklabels=['False', 'True'])
    plt.title(f'Confusion Matrix for {method_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python


# Affichage des matrices de confusion pour KMP et DFA
plot_confusion_matrix(confusion_kmp, 'KMP')
plot_confusion_matrix(confusion_dfa, 'DFA')

confusion_kmp, confusion_dfa
```

# Total time to construct DFA/LPS and find pattern

```python
df['Pattern_Label'] = ['m' + str(i+1) for i in range(len(df))]



# Courbe des temps avec les labels modifi√©s

plt.figure(figsize=(10, 6))

plt.plot(df['Pattern_Label'], df['time_kmp'], marker='o', label='KMP', color='b')

plt.plot(df['Pattern_Label'], df['time_dfa'], marker='o', label='DFA', color='r')

plt.title("Comparaison des temps d'ex√©cution (KMP vs DFA)")

plt.xlabel("Pattern")

plt.ylabel("Temps (Œºs)")

plt.legend()

plt.grid(True)

plt.show()



# Nuage de points des temps avec les labels modifi√©s

plt.figure(figsize=(10, 6))

plt.scatter(df['Pattern_Label'], df['time_kmp'], color='b', label='KMP')

plt.scatter(df['Pattern_Label'], df['time_dfa'], color='r', label='DFA')

plt.title("Nuage de points des temps d'ex√©cution (KMP vs DFA)")

plt.xlabel("Pattern")

plt.ylabel("Temps (Œºs)")

plt.legend()

plt.grid(True)

plt.show()
```

# Time to construct

```python
df_stats = pd.read_csv("../results/speed.csv")
```

```python


#! Enlever outliers DFA
# Labels pour les m√©thodes
max_dfa_avg = df_stats["MAX_DFA"].mean()
# D√©finition du seuil : 80 % de la moyenne de la colonne MAX_DFA
threshold = max_dfa_avg * 0.8

# Suppression des lignes o√π MAX_DFA d√©passe 80 % de la moyenne de MAX_DFA
df_stats = df_stats[df_stats["MAX_DFA"] <= threshold]

#! Enlever outliers KMP
max_kmp_avg = df_stats["MAX_KMP"].mean()
# D√©finition du seuil : 80 % de la moyenne de la colonne MAX_DFA
threshold = max_kmp_avg * 0.8
# Suppression des lignes o√π MAX_DFA d√©passe 80 % de la moyenne de MAX_DFA
df_stats = df_stats[df_stats["MAX_KMP"] <= threshold]



methods = ['m' + str(i+1) for i in range(len(df_stats))]

df_stats['Method_Label'] = methods



# Plot pour les valeurs MAX

plt.figure(figsize=(10, 6))

plt.plot(df_stats['Method_Label'], df_stats['MAX_KMP'], marker='o', label='MAX KMP', color='b')

plt.plot(df_stats['Method_Label'], df_stats['MAX_DFA'], marker='o', label='MAX DFA', color='r')

plt.title("Valeurs MAX (KMP vs DFA)")

plt.xlabel("Method")

plt.ylabel("Valeur (Œºs)")

plt.legend()

plt.grid(True)

plt.show()



# Plot pour les valeurs AVG

plt.figure(figsize=(10, 6))

plt.plot(df_stats['Method_Label'], df_stats['AVG_KMP'], marker='o', label='AVG KMP', color='b')

plt.plot(df_stats['Method_Label'], df_stats['AVG_DFA'], marker='o', label='AVG DFA', color='r')

plt.title("Valeurs AVG (KMP vs DFA)")

plt.xlabel("Method")

plt.ylabel("Valeur (Œºs)")

plt.legend()

plt.grid(True)

plt.show()



# Plot pour les valeurs MIN

plt.figure(figsize=(10, 6))

plt.plot(df_stats['Method_Label'], df_stats['MIN_KMP'], marker='o', label='MIN KMP', color='b')

plt.plot(df_stats['Method_Label'], df_stats['MIN_DFA'], marker='o', label='MIN DFA', color='r')

plt.title("Valeurs MIN (KMP vs DFA)")

plt.xlabel("Method")

plt.ylabel("Valeur (Œºs)")

plt.legend()

plt.grid(True)

plt.show()
```

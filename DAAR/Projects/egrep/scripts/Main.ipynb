```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns 
import pandas as pd 
```

# Accuracy 

```python
#Cr√©ation du DataFrame
df = pd.read_csv('../results/accuracy_and_time.csv')

# Matrices de confusion pour KMP et DFA
confusion_kmp = confusion_matrix(df['appears_expected'], df['found_kmp'])
confusion_dfa = confusion_matrix(df['appears_expected'], df['found_dfa'])
```

```python
# Fonction pour afficher les matrices de confusion
def plot_confusion_matrix(conf_matrix, method_name):
    plt.figure(figsize=(6, 4))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=['False', 'True'], yticklabels=['False', 'True'])
    plt.title(f'Confusion Matrix for {method_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Affichage des matrices de confusion pour KMP et DFA
plot_confusion_matrix(confusion_kmp, 'KMP')
plot_confusion_matrix(confusion_dfa, 'DFA')

confusion_kmp, confusion_dfa
```

# Time of finding

```python
df['Pattern_Label'] = ['m' + str(i+1) for i in range(len(df))]



# Courbe des temps avec les labels modifi√©s

plt.figure(figsize=(10, 6))

plt.plot(df['Pattern_Label'], df['time_kmp'], marker='o', label='KMP', color='b')

plt.plot(df['Pattern_Label'], df['time_dfa'], marker='o', label='DFA', color='r')

plt.title("Comparaison des temps d'ex√©cution (KMP vs DFA)")

plt.xlabel("Pattern")

plt.ylabel("Temps (ms)")

plt.legend()

plt.grid(True)

plt.show()



# Nuage de points des temps avec les labels modifi√©s

plt.figure(figsize=(10, 6))

plt.scatter(df['Pattern_Label'], df['time_kmp'], color='b', label='KMP')

plt.scatter(df['Pattern_Label'], df['time_dfa'], color='r', label='DFA')

plt.title("Nuage de points des temps d'ex√©cution (KMP vs DFA)")

plt.xlabel("Pattern")

plt.ylabel("Temps (ms)")

plt.legend()

plt.grid(True)

plt.show()
```
